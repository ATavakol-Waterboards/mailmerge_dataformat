```python
#### mapping datafilsa field column numbers to refdatafilsa field/column numbers
indxmap=np.ones((refdata_filsa.shape[0]))*-999
for i in np.arange(0,refdata_filsa.shape[0]):
    if refdata_filsa[i] in data_filsa:
        #print(refdata_filsa[i],np.where(data_filsa==refdata_filsa[i])[0][0])
        indxmap[i]=np.int(np.where(data_filsa==refdata_filsa[i])[0][0])
        
indxmap[indxmap==-999]=np.nan
apnindex=np.where(data_filsa=='APN')[0][0]
```

```python
ref_firstname_colno=np.where(refdata_filsa=='OWNER_FIRST')[0][0]
ref_lastname_colno=np.where(refdata_filsa=='OWNER_LAST')[0][0]
_firstname_colno=np.where(data_filsa=='FirstName')[0][0]
_lastname_colno=np.where(data_filsa=='LastName')[0][0]
print(ref_firstname_colno,ref_lastname_colno,_firstname_colno,_lastname_colno)
indxmap[ref_firstname_colno]=_firstname_colno
indxmap[ref_lastname_colno]=_lastname_colno
```

```python
##### creates list of headers including apns and creates an empty array/dataframe
#del(corrcode_unqs_n)
corrcode_data_ix=np.where(data_filsa=='CorrespondenceCode')[0][0]
corrcode_unqs=np.unique(dataa[:,corrcode_data_ix])
corrcode_unqs_n=corrcode_unqs.shape[0]
idxs=[]
noapns=[]
for i in np.arange(0,corrcode_unqs_n):
    idxs.append(np.str(i))
    noapns.append(len(np.where(dataa[:,corrcode_data_ix]==corrcode_unqs[i])[0])) 
maxnoapns=np.nanmax(noapns)  ##### max number of apns per corr-code from list of APN # fields
apnlist=[]
for i in np.arange(0,maxnoapns):
    apnlist.append('APN '+np.str(i+1))
stdfilsix=(np.where(refdata_filsa=='APN 1')[0][0])-1
newfilslist=np.concatenate((refdata_filsa[0:stdfilsix],apnlist))
print(refdata_filsa[0:stdfilsix])
newdf=pd.DataFrame(columns=newfilslist,index=idxs)

newdf
```

```python
#### filling the array or dataframe
for i in np.arange(0,corrcode_unqs_n):
    #idxs.append(np.str(i))
    tixs=np.array(np.where(dataa[:,corrcode_data_ix]==corrcode_unqs[i])[0])
    print(tixs,tixs.shape[0])
    for j in np.arange(0,newfilslist.shape[0]):
        #print(newfilslist[j])
        if np.isnan(indxmap[j])==0:
            #print(newdf.loc[np.str(i),newfilslist[j]],dataa[tixs[0],np.int(indxmap[j])])
            newdf.at[np.str(i),newfilslist[j]]=dataa[tixs[0],np.int(indxmap[j])]
            #print(newdf.loc[np.str(i),newfilslist[j]])
        ### find matching field/col-name from datasource and put value in newdf with newdf.at[ind,col]
    for k in np.arange(0,tixs.shape[0]):
        #print(np.str(i),np.str('APN '+np.str(k)))
        newdf.at[np.str(i),np.str('APN '+np.str(k+1))]=dataa[tixs[k],apnindex]
        #print(tixs[k])
        #print(np.where(dataa[:,corrcode_data_ix]==corrcode_unqs[i])[0])
```
